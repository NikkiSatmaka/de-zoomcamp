```python
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql://nikki:password@localhost:5433/ny_taxi")
```

```python
print(pd.io.sql.get_schema(df.to_pandas(), name="yellow_taxi_data", con=engine))
```

```python
lf = pl.scan_csv("yellow_tripdata_2021-01.csv", try_parse_dates=True)
```

```python
n_records = lf.select(pl.len()).collect().item()
print(n_records)
```

```python
lf.head().collect()
```

```python
lf.slice(0, 0).collect().write_database(
    table_name="yellow_taxi_data",
    connection=engine,
    if_table_exists="replace",
    engine="sqlalchemy",
)
```

```python
from time import time
```

```python
batch_size = 10_000
for i in range(0, n_records, batch_size):
    t_start = time()

    lf.slice(i, i + batch_size).collect().write_database(
        table_name="yellow_taxi_data",
        connection=engine,
        engine="sqlalchemy",
        if_table_exists="append",
    )

    t_end = time()

    print(f"inserted another chunk, took {t_end - t_start:.3f} seconds")
```
